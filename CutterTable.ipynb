```python
#Rule of Cutter Table
vowel = ["A","E","I","O","U"]

# After initial vowels
ruleVowel = {}
for ruleVowelKey in ["A","B","C"]:
    ruleVowel[ruleVowelKey] = 2
for ruleVowelKey in ["D","E","F","G","H","I","J","K"]:
    ruleVowel[ruleVowelKey] = 3
for ruleVowelKey in ["L","M"]:
    ruleVowel[ruleVowelKey] = 4
for ruleVowelKey in ["N","O"]:
    ruleVowel[ruleVowelKey] = 5
for ruleVowelKey in ["P","Q"]:
    ruleVowel[ruleVowelKey] = 6
for ruleVowelKey in ["R"]:
    ruleVowel[ruleVowelKey] = 7
for ruleVowelKey in ["S","T"]:
    ruleVowel[ruleVowelKey] = 8
for ruleVowelKey in ["U","V","W","X","Y","Z"]:
    ruleVowel[ruleVowelKey] = 9

# After initial letter S
ruleS = {"A":2,
        "C":3,
        "E":4,
        "H":5, "I":5,
        "M":6, "N":6, "O":6, "P":6,
        "T":7,
        "U":8,
        "W":9, "X":9, "y":9, "Z":9}

# After initial letters Qu
ruleQu = {"A":3,
        "E":4,
        "I":5,
        "O":6,
        "R":7,
        "T":8,
        "Y":9}

# For initial letters Qa-Qt, use 2-29, be assigned by local repository.
# This calculation is just for reference.
ruleQa = {"A":2,"B":10,"C":11,"D":12,"E":13,"F":14,
        "G":15,"H":16,"I":17,"J":18,"K":19,"L":20,
        "M":21,"N":22,"O":23,"P":24,"R":25,"S":26,"T":27}

# After other initial consonants
ruleCon = {}
for ruleConKey in ["A","B","C","D"]:
    ruleCon[ruleConKey] = 3
for ruleConKey in ["E","F","G","H"]:
    ruleCon[ruleConKey] = 4
for ruleConKey in ["I","J","K","L","M","N"]:
    ruleCon[ruleConKey] = 5
for ruleConKey in ["O","P","Q"]:
    ruleCon[ruleConKey] = 6
for ruleConKey in ["R","S","T"]:
    ruleCon[ruleConKey] = 7
for ruleConKey in ["U","V","W","X"]:
    ruleCon[ruleConKey] = 8
for ruleConKey in ["Y","Z"]:
    ruleCon[ruleConKey] = 9

# For expansion
ruleExp = {}
for ruleExpKey in ["A","B","C","D"]:
    ruleExp[ruleExpKey] = 3
for ruleExpKey in ["E","F","G","H"]:
    ruleExp[ruleExpKey] = 4
for ruleExpKey in ["I","J","K","L"]:
    ruleExp[ruleExpKey] = 5
for ruleExpKey in ["M","N","O"]:
    ruleExp[ruleExpKey] = 6
for ruleExpKey in ["P","Q","R","S"]:
    ruleExp[ruleExpKey] = 7
for ruleExpKey in ["T","U","V"]:
    ruleExp[ruleExpKey] = 8
for ruleExpKey in ["W","X","Y","Z"]:
    ruleExp[ruleExpKey] = 9

restart = "Y"
while restart != "N":

    cutter = ""
    inputWord = input("Enter word: ")
    inputWord = inputWord.upper()
    head = inputWord[0]  # First letter of the Cutter number
    cutter += head

    # Base on the initial letter of the last name entered, define the first number: 
    if head is "Q":
        if inputWord[1] is "U":
            cutter += str(ruleQu[inputWord[2]])
        else:
            print("Qa-Qt results should only be used as reference")
            cutter += str(ruleQa[inputWord[1]])
    elif head in vowel:
        cutter += str(ruleVowel[inputWord[1]])
    elif head is "S":
        cutter += str(ruleS[inputWord[1]])
    elif head not in vowel:
        cutter += str(ruleCon[inputWord[1]])
    
    # Define the thrid character and beyond
    if len(inputWord) >= 2:
        if head is "Q" and inputWord[1] is "U":
            i = 3
        else:
            i = 2
        while i <= len(inputWord)-1:
            cutter += str(ruleExp[inputWord[i]])
            i += 1
    
    print(cutter)
    restart = input("Cutter another word? (Y/N) ").upper()
```
